```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
producer = KafkaProducer(bootstrap_servers=['YOUR_PUBLIC_IP:9092'],
                         value_serializer=lambda x:
                         dumps(x).encode('utf-8'))
```

```python
producer.send('topic-name', value={'hello':'world'})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(10, 6))
plt.hist(df['Date'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution des Dates")
plt.xlabel("Date")
plt.ylabel("Fréquence")
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Date'], color='purple')
plt.title("Distribution de la densité des Dates")
plt.xlabel("Date")
plt.ylabel("Densité")
plt.xticks(rotation=45)
plt.show()
```

```python
while True:
    dict_stock = df.sample(1).to_dict(orient="records")[0]
    producer.send('topic-name', value=dict_stock)
    sleep(1)
```
